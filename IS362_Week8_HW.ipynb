```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Load data

df = pd.read_fwf('https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data', header=None)
df.head()
```

```python
# Change column names

df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight',
                'acceleration', 'model_year', 'origin', 'car_name']
df
```

```python
# Replace ? in horsepower column to a missing value

df.horsepower.replace('?', np.nan, inplace=True)
```

```python
df
```

```python
# Check the data type of each column

df.info()
```

```python
# Change data type of horsepower to a float

df['horsepower'] = df.horsepower.astype(float)

#Verify if changed successfully

df.info()
```

```python
# Replace origin's values

df.origin.replace(1, 'USA', inplace=True)
df.origin.replace(2, 'Asia', inplace=True)
df.origin.replace(3, 'Europe', inplace=True)
```

```python
df
```

```python
%matplotlib inline
```

```python
# Create a bar chart that shows the distribution for cylinders.

df.groupby('cylinders')
df['cylinders'].plot(kind='hist')
```

```python
# Create a scatterplot that shows the relationship between horsepower and weight

plt.scatter(df.horsepower, df.weight)
plt.ylabel('Weight')
plt.xlabel('Horsepower')
```

```python
import seaborn as sns
sns.set(color_codes=True)
```

## What's the relationship between a car's weight and miles per gallon?

```python
sns.regplot(x='weight', y='mpg', data=df)
```

According to the graph, cars that weigh less have a higher miles per gallon compared to cars that weigh more.
